```python
import pandas as pd
import numpy as np

df = pd.read_csv("movies_info.csv")
print("Dataset shape:", df.shape)
print("Dataset info:")

# Displaying data types and basic information about the DataFrame
print(df.dtypes)

# Displaying the first few rows, last few rows, and descriptive statistics of the DataFrame
df.info()

# Displaying the first few rows, last few rows, and descriptive statistics
print(df.head())

# Displaying the last few rows and descriptive statistics
print(df.tail())

# Displaying descriptive statistics of the DataFrame
print(df.describe())

print("\nMissing Values:")
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
missing_df = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage': missing_percentage
})
print(missing_df[missing_df['Missing Values'] > 0])
```
